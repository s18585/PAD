3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

```python
import pandas as pd
import numpy as np

orders = pd.read_csv("/orders.csv")
orders
```

```python
# 1 Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
orders['order_date']
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'])
orders['order_date']
```

```python
# 2 Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
orders['tshirt_category'].nunique()
orders['tshirt_category'].unique()

print(str(orders['tshirt_category'].nunique()) + ' ||| ' + str(orders['tshirt_category'].unique()))
```

```python
# 3 Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;

orders['tshirt_category'].unique()
orders['tshirt_category'] = orders['tshirt_category'].apply(
    lambda x: x
    .replace('Tshirt ', 'T-Shirt ')
    .replace('Wh ', 'White ')
    .replace('Bl ', 'Black ')
  )

orders['tshirt_category'].unique()
```

```python
# 4 Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply. dodanie dodatkowych danych do wartości, których nie da się rozdzielic na 3 (Hoodie, Tennis Shirt)
orders['tshirt_category'] = orders['tshirt_category'].apply(
    lambda x: x
    .replace('Hoodie', 'Default Hoodie U')
    .replace('Tennis Shirt', 'Default Tennis Shirt U') 
  )

orders['tshirt_type'] = orders['tshirt_category'].apply(
    lambda x: ' ' .join(x
    .split(' ')[1:-1] 
    )
  )

orders['tshirt_size'] = orders['tshirt_category'].apply(
    lambda x: x
    .split(' ')[-1]
  )

orders['tshirt_colour'] = orders['tshirt_category'].apply(
    lambda x: x
    .split(' ')[0]
  )

orders['tshirt_category'].unique()
```
